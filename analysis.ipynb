```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_data.csv")
df.head()
df.info()
df.shape
df.columns
df.isnull().sum()
df.describe()
df.dropna() #df.fillna() #df.dropna(inplace=True)
print("Average Final Grade:", df["G3"].mean())
print("Highest Final Grade:", df["G3"].max())
print("Lowest Final Grade:", df["G3"].min())
df.groupby("sex")["G3"].mean()
df.groupby("sex")["G3"].mean().plot(kind='bar', color=['blue', 'pink'])
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.show()

df["G1"].mean()
df["G2"].max()
df["G3"].min()
df["absences"].mean()
plt.hist(df["G3"], bins=10, color='blue', edgecolor='black')
plt.title("Distribution of Final Grades (G3)")
plt.show()
plt.scatter(df["G1"], df["G3"], color='red')
plt.title("G1 vs G3")
plt.xlabel("G1")
plt.ylabel("G3")
plt.show()
df.groupby("sex")["G3"].mean().plot(kind='bar', color=['blue', 'pink'])
plt.title("Average Final Grade (G3) by gender")
plt.show()
plt.scatter(df["studytime"], df["G3"], color='green')
plt.title("Study Time vs Final Grade (G3)")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
plt.scatter(df["failures"], df["G3"], color='orange')
plt.title("Number of Failures vs Final Grade (G3)")
plt.xlabel("Number of Failures")
plt.ylabel("Final Grade (G3)")
plt.show()
plt.scatter(df["absences"], df["G3"], color='purple')
plt.title("Number of Absences vs Final Grade (G3)")
plt.xlabel("Number of Absences")
plt.ylabel("Final Grade (G3)")
plt.show()
df.groupby("internet")["G3"].mean()
df.groupby("internet")["G3"].mean().plot(kind='bar', color=['gray', 'cyan'])
plt.title("Average Final Grade (G3) by Internet Access")
plt.xlabel("Internet Access")
plt.ylabel("Average Final Grade (G3)")
plt.show()
df.groupby("famsup")["G3"].mean()
df.groupby("famsup")["G3"].mean().plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title("Average Final Grade (G3) by Family Support")
plt.xlabel("Family Support")
plt.ylabel("Average Final Grade (G3)")
plt.show()
#Correlation analysis
corr = df.corr(numeric_only=True)
corr["G3"].sort_values(ascending=False)

#Top Performers
top_performers =  df.sort_values("G3", ascending=False)
top_performers[["age","sex","studytime","failures","absences","G3"]].head(10)

#Low Performers
low_performers = df.sort_values("G3", ascending=True)
low_performers[["age","sex","studytime","failures","absences","G3"]].head(10)
plt.hist(df["G3"], bins=10, color='blue', edgecolor='black')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```
